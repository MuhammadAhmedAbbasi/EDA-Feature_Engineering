# In Data analysis what we do

1) Missing Value
2) Explore about numerical variable
3) Explore about categorical variable
4) Finding relation between features

```python
# Missing Value
# First way of checking Null Values
df.isnull().sum()
```

```python
# Missing Value
# Second way of checking Null Values
# List Comprehension
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
# Missing Value
# Third way of checking Null Values
# Heat Map
# The reason we cant see null is because they are very less

sns.heatmap(df.isnull(),yticklabels=False,cmap='viridis')
```

```python
# We will see other dataset which is country code 

df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df['Country Code']
```

```python
df_country
```

```python
# Now we have to merge df_country with df
# we use .merge command
# "on" in .erge command tell which feature of df I have to relate my new added feature and how= tell that which dataframe is more important

final_df=pd.merge(df,df_country,on='Country Code',how='left') 
```

```python
final_df.head(2)
```

```python
final_df.columns
```

```python
# Now I perform EDA with respect of country name, in which country zomato is in top

Country_name=final_df['Country'].value_counts().index
country_val=final_df['Country'].value_counts().values
```

```python
plt.pie(country_val[:3],labels=Country_name[:3],autopct='%1.2f%%')
```

### Observation


1) Most of the zomato buisness is in India 
2) USA and UK are on second and third place

```python
a=final_df[final_df['Country']=='India']
a['City'].value_counts().plot(kind='bar')
```

### Observation 

In India most of the revenue is generated in New Dehli

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().sort_index(ascending=True).rename(columns={0:'Rating Count'}).plot(kind='bar')
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().sort_index(ascending=True).rename(columns={0:'Rating Count'})
ratings
```

```python
ratings_country=final_df.groupby(['Aggregate rating','Rating color','Rating text','Country','City']).size().reset_index().sort_index(ascending=True).rename(columns={0:'Rating Count'})
ratings_country[ratings_country['Country']=='India']
```

```python
plt.figure(figsize=(13, 5))
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['black','red','orange','yellow','green','green'])
```

### Observation


1) Not rated count is very high
2) Maximum number of rating is between 2.5 to 3.6

```python
sns.countplot(x='Rating color',data=ratings,palette=['black','red','orange','yellow','green','green'])
```

```python
# Find the countries name that has given zero rating

ratings_country=final_df.groupby(['Aggregate rating','Rating color','Rating text','Country']).size().reset_index().sort_index(ascending=True).rename(columns={0:'Rating Count'})
ratings_country[ratings_country['Aggregate rating']==0.0].sort_index(ascending=True)
```
