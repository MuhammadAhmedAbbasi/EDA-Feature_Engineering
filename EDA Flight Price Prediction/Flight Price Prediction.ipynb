```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_Train.xlsx')
test_df=pd.read_excel('Test_set.xlsx')
```

```python
train_df
```

```python
df=pd.merge(train_df,test_df,how='outer')
df
```

First we take Date_of_Journey, Here I have day, month and year. So we have already checked that it is in object dtype so first step is to convert into DateTime and then make seperate columns for day, month and year

```python
df.dropna(subset=['Route'], inplace=True)
df.dropna(subset=['Total_Stops'], inplace=True)
```

```python
df['Duration_hour']=(df['Duration'].str.split(' ').str[0]).str.split('h').str[0]
df.drop(6681,axis=0,inplace=True)
df.drop(13141,axis=0,inplace=True)
df['Duration_hour']=df['Duration_hour'].astype(int)
df['Duration_minute']=((df['Duration'].str.split(' ').str[1]).str.split('m').str[0]).fillna(0).astype(int)
df['Duration']=df['Duration_hour']*60+df['Duration_minute']
```

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_Minute']=df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
df['Departure_Hour']=df['Dep_Time'].str.split(':').str[0].astype(int)
df['Departure_Minute']=df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
# Feture Engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month']=df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year']=df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
source=df['Source'].value_counts().index
label1={}
for i,k in enumerate (source,0):
    label1[k]=i

df['Source']=df['Source'].map(label1)
```

```python
source2=df['Destination'].value_counts().index
label2={}
for i,k in enumerate (source2,0):
    label2[k]=i

df['Destination']=df['Destination'].map(label2)
```

```python
source3=df['Airline'].value_counts().index
label3={}
for i,k in enumerate (source3,0):
    label3[k]=i

df['Airline']=df['Airline'].map(label3)

```

```python
df['Total_Stops'].unique()


label4={
    'non-stop': 0,
    '1 stop':1,
    '2 stops':2,
    '3 stops':3,
    '4 stops':4,
}
df['Total_Stops']=df['Total_Stops'].map(label4)
```

```python
info=df['Additional_Info'].value_counts().sort_values().index
label5={}

for i,k in enumerate(info,0):
    label5[k]=i


df['Additional_Info']=df['Additional_Info'].map(label5)
```

```python
df.drop(['Duration_hour','Duration_minute','Arrival_Time','Dep_Time','Date_of_Journey','Route'],axis=1,inplace=True)
```

```python
df_train=df[~df['Price'].isnull()]
df_test=df[df['Price'].isnull()]
```
